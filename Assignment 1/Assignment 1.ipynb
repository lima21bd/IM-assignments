## Problem Description

The following equations need to be plotted in the same plot >>

1. Logarithmic = Θ(logn)
2. Linear = Θ(n)
3. Quadratic = Θ(n^2)
4. Cubic = Θ(n^3)
5. Polynomial = Θ(n^k)

### Input Values: n = 7, k = 4

*Author: Khadija Akter Lima*

```python
#Importing necessary libraries

import numpy as np
import matplotlib.pyplot as plt

import math
```

```python
x = np.array(list(range(1, 8)))
n = 7
k = 4

Logarithmic = np.log2(x)
Linear = x
Quadratic = x**2
Cubic = x**3
Plynomial= x**k

fig, ax = plt.subplots()
plt.rcParams['figure.figsize'] = [16, 10]


ax.plot(x, Logarithmic, label = 'Θ(logn)', marker = 'o')
ax.plot(x, Linear, label = 'Θ(n)', marker = 'o')
ax.plot(x, Quadratic, label = 'Θ(n^2)', marker = 'o')
ax.plot(x, Cubic, label = 'Θ(n^3)', marker = 'o')
ax.plot(x, Plynomial, label = 'Θ(n^k)', marker = 'o')


ax.set_xlabel('Input Size, n')
ax.set_ylabel('Computational Cost, Θ')

ax.legend();
```
